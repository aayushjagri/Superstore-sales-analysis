# **Most Selling And Profitable Category By Sub-Category**

```python
df.columns
```

```python
category_df = df.groupby(['category', 'sub_category'])['sales', 'profit'].mean().sort_values(by = 'category', ascending = False).reset_index()
```

```python
category_df
```

```python
#sales_per_category
sns.set_style('darkgrid')
plt.figure(figsize = (12,6))

sns.barplot(x = 'category',
            y =  'sales', 
            data = category_df, hue = 'sub_category')
plt.legend(fontsize = 8)
plt.title(' Total Sales Per Category By Sub-category');

```

# ****We can see from above chart that;****
* The Average most selling sub-category is Tables followed by Coplers & Machines in category Furniture, Office &  Technology respectively. 
* ****Does this means Tables is the highest profit earning Sub-Category? Let's find out;****

```python
#Profit_per_Category
plt.figure(figsize = (12,7))
sns.barplot(x = 'category',
            y =  'profit', 
            data = category_df, hue = 'sub_category')
plt.legend(fontsize = 7)
plt.title('Average Profit Per Category By Sub-category');
```

* Even though Average most selling sub-category is Table, looking at above chart we can see that Coplers is the highest average profit generator in any given sub-category.
Most of the losses are from tables  while furniture category items such as chairs, bookcase and furnishings have been profiting.

* ***We can see why the furniture category raising profit at small pace compare to other category because most of the losses are from tables while furniture category items such as chairs, bookcases and furnishings have been profiting.***

# **Regional Sales & Profit  Analysis**

```python
df.segment.unique()
```

```python
df1 = df.groupby(['region', 'category'])['sales'].sum().sort_values().unstack()
df2 = df.groupby(['region', 'category'])['profit'].mean().sort_values().unstack()

df3 = df.groupby(['region', 'segment'])['sales'].sum().sort_values().unstack()
df4 = df.groupby(['region', 'segment'])['profit'].sum().sort_values().unstack()

df5 = df.groupby(['region',  'year'])['sales'].sum().sort_values().unstack()
df6 = df.groupby(['region', 'year'])['profit'].sum().sort_values().unstack()
```

```python
fig, ([ax1, ax2], [ax3, ax4], [ax5, ax6]) = plt.subplots(nrows = 3, ncols = 2, figsize = (15, 10))

ax1 = sns.heatmap(df1, ax = ax1, annot = True,fmt = ".0f", cmap = 'cividis')
ax2 = sns.heatmap(df2, ax = ax2, annot = True,fmt = ".0f", cmap = 'cividis')
ax3 = sns.heatmap(df3, ax = ax3, annot = True,fmt = ".0f", cmap = 'rocket_r')
ax4 = sns.heatmap(df4, ax = ax4, annot = True,fmt = ".0f", cmap = 'rocket_r')
ax5 = sns.heatmap(df5, ax = ax5, annot = True,fmt = ".0f", cmap = 'magma_r')
ax6 = sns.heatmap(df6, ax = ax6, annot = True,fmt = ".0f", cmap = 'magma_r')
ax1.title.set_text('Fig 1. Sales Per Category by Region')
ax2.title.set_text('Fig 1.1. Profit Per Category By Region')
ax3.title.set_text('Fig 2. Sales Per Segment By Region')
ax4.title.set_text('Fig 2.1. Profit Per Segment By Region')
ax5.title.set_text('Fig 3. Sales Per Region (Each Year)')
ax6.title.set_text('Fig 3.1. Profit  Per Region(EachYear)')
#fig.subplots_adjust(wspace=0.3)
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

**We can see from the above charts that;**
* In Fig 1 & 1.1 The total most Selling & Profitable category by region is Technology followed by Office Supplies and Furniture in Central Region.  
* In Fig 2 & 2.1 The total most Selling & Profitable Segment by region is Consumer followed by Corporate  and Home Office in Central Region.  
* In Fig 3 & 3.1 The total most Selling & Profitable Year per region is 2014 followed by 2013, 2012  and 2011 in Central Region.
